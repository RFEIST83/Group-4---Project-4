```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
array_streamingdata = []
array_num = np.arange(0,500,1, dtype = int)
for doc in TS_sd.find({}):
    #pprint.pprint(doc)
    array_streamingdata.append(doc)
TS_streamingdata_df = pd.DataFrame(data = array_streamingdata, index = np.arange(0, 500, 1, dtype = int) )
TS_streamingdata_df.convert_dtypes(infer_objects = False, convert_string = True)
New_TS_Streamingdata=TS_streamingdata_df[["Song Title", "Total Streams", "Daily Streams"]]
New_TS_Streamingdata = New_TS_Streamingdata.convert_dtypes(infer_objects = True, convert_string = True)
New_TS_Streamingdata.dtypes
```

```python
tstracks = []
for coll in TS_tracks.find({}):
    tstracks.append(coll)
TS_trackinfo_df = pd.DataFrame(data=tstracks, index = np.arange(0, 246, 1, dtype=int))
TS_trackinfo_df.astype({"album_id": 'string', "album_name":'string', "name": 'string'}).dtypes
new_TS_trackinfo = TS_trackinfo_df[["album_name", "name","release_date","duration","track_number","popularity","explicit","uri"]]
new_TS_trackinfo
v = pd.Series(new_TS_trackinfo.groupby('album_name')["popularity"].sum())
c = pd.Series(new_TS_trackinfo.groupby('album_name')['name'].count())
y = c.values.reshape(-1, 1)
#print(f"the number of songs in each album: \n{c}")
x = v.values.reshape(-1, 1)
#create a model with scikit-learn
model = LinearRegression()
#fit the data into the model
model.fit(x, y)
#display the slope
print(f"model's slope: {model.coef_}")
#display the y-intercept
print(f"model's y-intercept : {model.intercept_}")
print(f"model's formula: y = {model.intercept_}+{model.coef_[0]}x")
# the minimum release year being 2006 we are using the x value as 18 2006 - 2024 the albums were released the value is used for prediction of trending popularity.
print(f"model's formula: y= {model.intercept_}+{model.coef_[0]}*18")
#predicting rate of popularity increase
y_18 = model.intercept_+model.coef_[0]* 18
#display prediction
print(f"prediction for popularity with a span of 18 years: {y_18}")
predicted_y_values = model.predict(x)
df_popularity = pd.DataFrame(data = v)
df_popularity["predicted popularity"] = predicted_y_values
# for each index being a unique album name
df_popularity['album_name'] = new_TS_trackinfo['album_name'].unique()
df_popularity
df_popularity.set_index(np.arange(0, df_popularity['popularity'].size, 1))
Albumname = df_popularity['album_name'].array
#creating a shortened version of album name to make it easy to plot xaxis
x_label = []
for i in range(len(Albumname)):
    b = Albumname[i].split()
    x_label.append(b[0])
x_label
df_popularity['X_Label'] = x_label
df_popularity
```

```python
import hvplot.pandas
popularity_plot = df_popularity.hvplot.scatter(x = "X_Label", y = "predicted popularity", rot = 90, title ="predicted increase in popularity in each of the albums")
popularity_plot
```

```python
from sklearn.metrics import mean_squared_error, r2_score
x = df_popularity['popularity'].values.reshape(-1,1)
y = df_popularity['predicted popularity'].values.reshape(-1,1)

score = model.score(x, y, sample_weight = None)
r2 = r2_score(y, predicted_y_values)
mse = mean_squared_error(y, predicted_y_values)
rmse = np.sqrt(mse)
std = np.std(y)
print(f"The score is :{score}")
print(f"The r2 score is :{r2}")
print(f"The mean squared error is :{mse}")
print(f"The root mean squared error is :{rmse}")
print(f"The standard deviation is : {std}")
```
